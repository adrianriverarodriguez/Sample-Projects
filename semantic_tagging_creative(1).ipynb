```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

# Theme reference texts
housing_text = """
Housing and Community Development— Appeared in over 40% of Creative Solutions responses and reflects a strong interest
in aligning school facility decisions with broader neighborhood revitalization goals. Respondents proposed using district-owned land for
affordable housing development to attract and retain families in the district. This approach was framed not only as a solution to declining
enrollment but also as an opportunity to reinvest in Evanston’s communities by addressing housing access, strengthening neighborhood stability,
and ultimately supporting school sustainability through population growth. Responses also highlighted historical inequities that should be
considered should the district use this opportunity to reinvest in neighborhood and community revitalization and suggest stronger partnerships
with anchor institutions to support local revitalization.


Example quotes:
- “Fundraising and increasing revenues like the City did to build Robert Crown and D202 did for their new wing… Look to raise fees across the board (and continue to offer financial assistance when needed).”
- “Selling plots of land to build housing for families; doing so would also increase the number of families available to enroll in the district.”
- “I think the district needs to ask more of Northwestern, which doesn’t pay property tax and could contribute more to the education of the community it resides in.”
- “Equitable financial support of schools based upon racial mix… With largely white schools in North, the lion’s share of these funds would properly end up in the south.
"""

revenue_text = """
Revenue-Generating Ideas—Revenue-generating ideas were shared in over a third of all Creative Solutions responses, centered on ways the district
could generate income through strategic use of its existing assets rather than closing or selling school properties outright. These ideas reflect
a pragmatic and resourceful mindset among respondents, emphasizing sustainability and long-term investment over short-term cost-cutting. Respondents
proposed leasing school spaces to community organizations, early childhood providers, and after-school programs, as well as renting facilities for evening or weekend use.
Others suggested seeking partnerships with local businesses, launching fundraising campaigns, or exploring referenda as ways to bolster district finances
Open enrollment was also framed as a revenue generating idea that would provide parents with school choice while generating money for the district


Example quotes:
- “Consider how to use district resources, such as school buildings, as a source of income, e.g., leasing out spaces to other community orgs, early childhood programs, etc.”
- “Neighborhood schools could be rented out for private evening, weekend, or summer programming or camps.”
- “See if local businesses will donate services needed to repair run-down facilities.”
- “I would also consider not selling the building that will be closed immediately. There might be leases to community organizations, the city, etc., that could be explored to continue to use the building as a possible resource or investment.”
- "Allowing open enrollment and allowing parents to apply to school of their choice—if it’s not in their boundary, they can pay a small fee to attend school of choice.”
"""

repurposing_text = """
Repurposing School Buildings—Thirty percent of responses expressed strong interest in keeping school properties within district control and ownership. These responses were often driven by concerns that selling buildings now would limit the
district’s capacity to support future enrollment growth. Instead of selling, respondents advocated for retaining properties and repurposing them to meet evolving community needs—such as
leasing space for early childhood programs, community centers, or summer camps. Repurposing was also proposed as a strategy to generate revenue while maintaining public stewardship of
district assets, aligning with a broader commitment to community-centered planning.

Example quotes:
- “Rent out district facilities; partner with community theaters, summer camps, etc. to host their activities and garner revenue.”
- “Repurpose buildings such as open Early Childhood/Childcare center or rent the space to childcare providers such as Kinder Care, Bright Horizons, or Goddard School.”
- “Open every building for summer camp revenue (i.e. Willard used to host camp inside the building, then that has not been allowed in almost 10 years). Do not sell property, lease it out.”
- “Transform parts of school buildings into community centers, co-working spaces, or incubators for local entrepreneurs… Partner with the city and university to fill vacant commercial spaces with education related ventures—such as early childhood centers, tutoring hubs, or special needs facilities…”
"""

childcare_text = """
Childcare and Early Learning Centers-Appeared in 21.4% of responses, highlighting the communities’ support for using school facilities to expand access to early childhood education
and childcare services. Many respondents proposed repurposing closed or underutilized buildings to house preschool programs, infant and toddler care, or partnerships
with private providers such as Bright Horizons or KinderCare. These ideas were seen as both a practical solution to meet the needs of working families and a strategic
investment in attracting and retaining families within the district. Several responses also noted the potential for onsite childcare to support district staff and offer
learning opportunities for older students interested in education or service roles


Example quotes:
-"If schools must be closed, retaining the buildings and creating top-notch preschool options at these buildings to both grow the attractiveness of the district and retain the properties should enrollment rebound”
-“Could you lease parts of an underutilized school like Orrington to a private preschool, for instance?”
-“Could schools be rented out to childcare providers? This could be beneficial for staff with infants and young children by offering another childcare option that would be convenient.”
-“Creating top-notch preschool options at these buildings to grow the attractiveness of the district.”
"""

cost_text = """
Cost Reduction Suggestions-As many as 17.9% of responses proposed that the district reduces other spending streams before resorting to school closures or property sales.
Examples included focusing on cutting administrative costs, limiting technology-related expenses, and improving energy efficiency.
Several participants called for reallocating funds toward student services and school maintenance rather than new infrastructure.

Example quotes:
-“Cut more expenses at central administration of District 65; District 65 still has too many staff employees rather than teachers.”
-“Maintain closed schools for community centers. Segment schools to close portions of them. Use energy as a service approach to cut HVAC and lighting costs.”
-“Redrawing the enrollment map so that every child can walk to a school close to their home would be a good place to start. The buildings already exist. Perhaps they need updating, and money should be allocated for this rather than new builds.”
-“Get rid of the 1-1 technology to save money. Keep the older school buildings open.”
-“Do all K/1 students require 1-1 iPads? How are subscriptions being used? Are all of them necessary?”
"""

inovation_text = """
Instructional Model Innovations- Some responses (5.7%) made calls to rethink traditional grade configurations to improve resource use,
educational outcomes, and neighborhood cohesion. Respondents suggested transitioning from the current elementary–middle school split to models such as K–8 schools or K–2 / 3–5 grade spans.
These alternative configurations were seen as opportunities to reduce transitions for students, better target instructional resources, and preserve walkable, neighborhood-based schooling.
In some cases, innovative models were also linked to cost savings through shared staffing, staggered schedules, or more efficient building utilization

Example quotes:
-“Including Pre-school in a larger combined school building and potentially looking at K-8 buildings.”
-“Have we ever considered K-2, 3-5 schools? Or other alternative configurations that may allow for more focused interventions and expertise in one building?”
-“Could there be K-2 and 3-5 schools? That way specific resources would not need to be shared across all buildings.”
-“Alternative funding sources should be considered including private donations. I also believe some of the elementary schools with extra capacity should be considered to be converted to K-8 schools, and consideration given to closing middle schools. This can help preserve walkability / neighborhood schools.”
"""

# Encode reference themes
house_emb = model.encode(housing_text, convert_to_tensor=True)
revenue_emb = model.encode(revenue_text, convert_to_tensor=True)
repurposing_emb = model.encode(repurposing_text, convert_to_tensor=True)
childcare_emb = model.encode(childcare_text, convert_to_tensor=True)
cost_emb = model.encode(cost_text, convert_to_tensor=True)
innovation_emb = model.encode(inovation_text, convert_to_tensor=True)
```

```python
def semantic_tag(response):
    resp_emb = model.encode(str(response), convert_to_tensor=True)
    sim_h = util.pytorch_cos_sim(resp_emb, house_emb).item()
    sim_r = util.pytorch_cos_sim(resp_emb, revenue_emb).item()
    sim_re = util.pytorch_cos_sim(resp_emb, repurposing_emb).item()
    sim_c = util.pytorch_cos_sim(resp_emb, childcare_emb).item()
    sim_co = util.pytorch_cos_sim(resp_emb, cost_emb).item()
    sim_i = util.pytorch_cos_sim(resp_emb, innovation_emb).item()
    tags = []
    if sim_h > 0.6:
        tags.append("Housing")
    if sim_r > 0.6:
        tags.append("revenue")
    if sim_re > 0.6:
        tags.append("repurposing")
    if sim_c > 0.6:
        tags.append("childcare")
    if sim_co > 0.6:
        tags.append("cost")
    if sim_i > 0.6:
        tags.append("innovation")
    return tags

# Apply to your dataframe
df["Matched Themes"] = df["Responses"].apply(semantic_tag)
df[["Responses", "Matched Themes"]]
```
